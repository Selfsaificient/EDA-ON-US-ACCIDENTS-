### START TIME 

```python
Accidents_df.Start_Time
```

```python
Accidents_df.Start_Time = pd.to_datetime( Accidents_df.Start_Time )
```

```python
Accidents_df.Start_Time.dt.hour 
```

```python

sns.distplot(  Accidents_df.Start_Time.dt.hour  , bins= 24, kde=False,  norm_hist=True  )
```

##### *THIS GRAPH IS SPECIFICALLY DEPICTING THE TIME PERIOD WHEN THE ACCIDENTS OCCURED.*

- A HIGH PERCENTAGE OF ACCIDENTS OCCUR BETWEEN 1 PM TO 5 PM 
- NEXT HIGHEST PERCENTAGE OF ACCIDENTS OCCUR BETWEEN 6 AM TO 8 AM ( PROBABLY DUE TO PEOPLE ARE IN A HURRY TO GET TO WORK)

```python
sns.set_style( 'white')
sns.distplot(  Accidents_df.Start_Time.dt.dayofweek  , bins= 7, kde=False,  norm_hist=True  )
```

#### *THIS GRAPH SPECIFICALLY SHOWS THE RANGE OF ACCIDENTS OCCURED IN A WEEK*

- On FRIDAY the number of accidents is the highest

- On Weekends the number of accidents is the lowest 

IS THE DISTRIBUTION BY HOUR THE SAME ON WEEKENDS AS ON WEEKDAYS ? 


```python
sundays_start_time = Accidents_df.Start_Time [ Accidents_df.Start_Time.dt.dayofweek == 6 ] 

sns.distplot(sundays_start_time.dt.hour , bins= 24, kde=False,  norm_hist=True  )
```

#### THIS GRAPH DEPICTS THE RANGE OF ACCIDENTS OCCURED ON SUNDAY AT DIFFERENT TIME  INTERVALS. 

```python
Monday_start_time = Accidents_df.Start_Time[ Accidents_df.Start_Time.dt.dayofweek == 0 ] 

sns.distplot(Monday_start_time.dt.hour , bins= 24, kde=False,  norm_hist=True  )
```

#### THIS GRAPH DEPICTS THE RANGE OF ACCIDENTS OCCURED ON MONDAY AT DIFFERENT TIME  INTERVALS. 

 BY THE ABOVE TWO GRAPHS WE  UNDERSATND THAT MORE ACCIDENTS OCCUR BFORE 5 AM  ON SUDNAYS AS COMPARED ON MONDAYS WHERE WE CAN SEE MORE ACCIDENTS OCCUR AFTER 5 AM

 
 ON SUNDAYS MOST OF THE PEOPLE HAVE OFF FROM THEIR WORK SO THEY CAN BE OUT TILL LATE THATS WHY WE SEE MORE ACCIDENTS BEFORE 5 AM WHEREAS ON MONDAYS WE KNOW PEOPLE HAVE TO RUSH TO GO TO  THEIR WORK SO WE SEE ACCIDENTS USUALLY AFTER 5 AM. 

```python
sns.distplot(  Accidents_df.Start_Time.dt.month   , bins= 12, kde=False,  norm_hist=True  )
```

#### THIS GRAPH DEPICTS THE OCCURENCE OF ACCIDENTS WITH RESPECT TO MONTHS OF ALL THE YEARS TOGETHER 

- WE CAN INFER FROM THIS GRAPH THAT DURING THE WINTER SEASON THERE IS A RISE IN THE NUMBER OF ACCIDENTS 
- MAXIMUM NUMBER OF ACCIDENTS OCCUR IN DECEMBER ( PROBABLY DUE TO SNOWFALL, FOG )

```python
df_2020 = Accidents_df[ Accidents_df.Start_Time.dt.year == 2020]
sns.distplot(  df_2020.Start_Time.dt.month   , bins= 12, kde=False,  norm_hist=True  )
```
